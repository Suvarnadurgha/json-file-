```python
# After uploading file to Lakehouse, read it into a Spark dataframe
# Make sure to update the path to match your environment
df = spark.read.json("Files/data/data_00.json")
# Display the dataframe content
display(df)
```

```python
#Print number of records and Schema
print(df.count())
df.printSchema()
```

```python
# Extract the 2 columns needed to solve Part 1 and Part 2
from pyspark.sql.functions import explode, col

accountsDF = df.select(explode(col("accountHistory")))
postsDF = df.select(explode(col("posts")))
display(accountsDF)
display(postsDF)
```

```python
from pyspark.sql.functions import sum, count
# Part 1 Solution
accountsDF.printSchema()
# Calculate Count 
part1DF = accountsDF.groupBy("col.type") \
    .agg(count("col.type").alias("Count"), \
         sum("col.amount").alias("Total_Amount")
     ) 
display(part1DF)
```
